## Halvorsen Attractor

La idea de este ejercicio es simular el movimiento de una partícula en un atractor extraño, su sistema de ecuaciones es el siguiente:<br>

\begin{equation} 
    \begin{cases}
        \dot{x} =  - \alpha x - 4y - 4z - y^2  \\
        \dot{y} =  -\alpha y - 4z - 4x - z^2 \\
        \dot{z} =  -\alpha z - 4x -4y - x^2  \\
    \end{cases}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#parámetros simulación
a= 1.89
#color de los ploteos
m= 'tomato'
#variables estáticas
dt= 0.01
tf= 100
t = np.arange(0 , tf + dt ,dt)

def Halvorsen(t,a,x_inicial,y_inicial,z_inicial):
    x = np.zeros(len(t))
    x1 = np.zeros(len(t)) #primer derivada
    y = np.zeros(len(t))
    y1 = np.zeros(len(t)) #primer derivada
    z = np.zeros(len(t))
    z1 = np.zeros(len(t)) #primer derivada
    #condiciones iniciales
    i=0
    x[i]=x_inicial
    y[i]=y_inicial  
    z[i]=z_inicial
    x1[i]=0
    y1[i]=0
    z1[i]=0
    
    while i < len(t) - 1:
        #método de Cromer, invierto el orden de cómputo, primero calculo la derivada (velocidad) y en base a ella calculo la nueva posicón. 
        x1[i+1] = -a*x[i] - 4*y[i] - 4*z[i] -y[i]**2
        y1[i+1] = -a*y[i] - 4*z[i] - 4*x[i] -z[i]**2
        z1[i+1] = -a*z[i] - 4*x[i] - 4*y[i] -x[i]**2
        x[i+1] = x[i] + x1[i+1] * dt
        y[i+1] = y[i] + y1[i+1] * dt
        z[i+1] = z[i] + z1[i+1] * dt
        i = i+1

    figure, axis = plt.subplots(1,2,figsize=(10,5),subplot_kw=dict(projection='3d'))
    axis[0].plot(x,y,z,color = 'tomato',alpha =0.7)
    axis[0].set_title(" x,y,z " + "a: " + str(a) )
    axis[1].plot(x1,y1,z1,color = 'teal',alpha =0.7)
    axis[1].set_title(" x',y',z' " + "a: " + str(a) )
    plt.show()

    figure, axis = plt.subplots(1,3,figsize=(15,5))
    axis[0].plot(x, y,color = 'tomato', alpha =0.7)
    axis[0].set_title("plano x y - " + "a: " + str(a) )
    axis[1].plot(y, z,color = 'tomato',alpha =0.7)
    axis[1].set_title("plano y z - " + "a: " + str(a) )
    axis[2].plot(x, z,color = 'tomato',alpha =0.7)
    axis[2].set_title("plano x z - " + "a: " + str(a) )
    plt.show()
    figure, axis = plt.subplots(1,3,figsize=(15,5))
    axis[0].plot(x1, y1, color = 'teal',alpha =0.7)
    axis[0].set_title("plano x' y' - " + "a: " + str(a) )
    axis[1].plot(y1, z1, color = 'teal',alpha =0.7 )
    axis[1].set_title("plano y' z' - " + "a: " + str(a) )
    axis[2].plot(x1, z1, color = 'teal',alpha =0.7 )
    axis[2].set_title("plano x' z' - " + "a: " + str(a) )
    plt.show()
    figure, axis = plt.subplots(1,3,figsize=(15,5))
    axis[0].plot(x1, x,color = 'green',alpha =0.7)
    axis[0].set_title("Diag.  de fase  x' x  - " + "a: " + str(a) )
    axis[1].plot(y1, y,color = 'green',alpha =0.7)
    axis[1].set_title("Diag. de fase  y' y  - " + "a: " + str(a) )
    axis[2].plot(z1, z,color = 'green',alpha =0.7)
    axis[2].set_title("Diag.  de fase  z' z  - " + "a: " + str(a) )
    plt.show()
    figure, axis = plt.subplots(1,3,figsize=(21,5))
    axis[0].plot(t, x,color = 'tomato',alpha =0.7)
    axis[0].set_title("evolucion t x - " + "a: " + str(a) )
    axis[1].plot(t, y,color = 'tomato',alpha =0.7)
    axis[1].set_title("evolucion t y - " + "a: " + str(a) )
    axis[2].plot(t, z,color = 'tomato',alpha =0.7)
    axis[2].set_title("evolucion t z - " + "a: " + str(a) )
    plt.show()
    figure, axis = plt.subplots(1,3,figsize=(21,5))
    axis[0].plot(t, x1,color = 'teal',alpha =0.7)
    axis[0].set_title("evolucion t x' - " + "a: " + str(a) )
    axis[1].plot(t, y1,color = 'teal',alpha =0.7)
    axis[1].set_title("evolucion t y' - " + "a: " + str(a) )
    axis[2].plot(t, z1,color = 'teal',alpha =0.7)
    axis[2].set_title("evolucion t z' - " + "a: " + str(a) )
    plt.show()
```
